### **题目描述**
### 利用散点图挖掘出“人力资源分析数据集”中员工对环境满意度和其每月加班平均时间的关系。

数据集中的字段包括：

    对环境的满意度(satisfaction_level)
    每月加班平均时间(average_montly_hours)


#### 1. 下载数据

```python
import os
import moxing as mox

if not os.path.exists("./HR_comma_sep.csv"):
    mox.file.copy("obs://modelarts-labs-bj4-v2/course/hwc_edu/python_module_framework/datasets/matplotlib/HR_comma_sep.csv", 
                  "./HR_comma_sep.csv")
```

#### 2. 读取数据并显示

```python
import pandas as pd
HR_data = pd.read_csv('./HR_comma_sep.csv')
HR_data.head()
```

#### 3. **代码实现**
请在下面 cell 中编写代码，实现题目描述的要求。

```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline

plt.rcParams['font.sans-serif'] =['SimHei']
plt.figure(figsize=(10,10))
plt.scatter(HR_data['average_montly_hours'], HR_data['satisfaction_level'])
plt.xlabel("每月加班平均时间")
plt.ylabel("对环境的满意度")
plt.title("员工对环境满意度和其每月加班平均时间的关系")
plt.show()
```
