#### 2. 具体功能实现

```python
# 导入模块
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv('./wechart.csv')
data['date']=pd.to_datetime(data['date'])
data.set_index('date',inplace=True)

plt.rcParams['font.sans-serif'] =['SimHei']
plt.figure(figsize=(10,10))
plt.xlabel("日期")
plt.ylabel("阅读人数")
plt.title("文章每天阅读的人数")
plt.plot(data["2017-8":].index, data["2017-8":]["article_reading_cnts"], '-o', markerfacecolor='r')
plt.show()

# 绘制效果示例：
```
