### **题目描述**
绘制一个函数

    在区间[0,2]上绘制函数f(x) = sin^2(x-2)*e^(-x^2)，并加上轴标签和标题。
    
    横轴：x
    纵轴：f(x)
    标题：f(x) = sin^2(x-2)*e^(-x^2)
提示：

    1.使用np.linspace来生成[0,2]区间任意数量的横轴坐标。
    2.使用np.sin和np.exp等函数来完成f(x)函数值的求解。
    3.使用plt.plot来作图。
    4.使用plt.xlabel, plt.ylabel来生成坐标轴标签，用plt.title来生成图像标题。

### **代码实现**
请在下面 cell 中编写代码，实现题目描述的要求。

```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline

"""
请在此处实现代码
"""
x = np.linspace(0,2,20, endpoint=True)
y = np.sin(x-2)**2*np.exp(-x**2)

plt.style.use("seaborn-darkgrid")
plt.plot(x,y, color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("f(x) = sin^2(x-2)*e^(-x^2)")
plt.show()
```
