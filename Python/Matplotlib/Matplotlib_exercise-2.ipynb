### **题目描述**
### 根据你对matplotlib的理解，请画出满足条件的直线y = x - 1。
1. 线型为虚线，线宽为1；
2. 纵坐标范围（-2，1），横坐标范围（-1，2），横纵坐标在（0，0）坐标点相交；
3. 横坐标的 [-1,-0.5,1] 分别对应 [bad, normal, good]。

### **代码实现**
请在下面 cell 中编写代码，实现题目描述的要求。

```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline

"""
请在此处实现代码
"""
x = np.arange(-1, 2, 0.1)
y = x-1

plt.figure(figsize=(10, 10))
plt.xlim(-1, 2)
plt.ylim(-2, 1)
plt.xticks([-1, -0.5, 1], ['bad', 'normal', 'good'])
plt.yticks(np.arange(-2, 1.5, 0.5))
plt.xlabel('X')
plt.ylabel('Y')
plt.title("y=x-1")

ax = plt.gca()
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

plt.plot(x, y, linewidth=1, color='red')
plt.show()
```
