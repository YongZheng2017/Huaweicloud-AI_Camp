### **题目描述**
### 使用matplotlib绘制3个子图：y = x; y = x^2; y = 0.01*x - 0.01,进行合并显示。
提示：plt.subplot()

### **代码实现**
请在下面 cell 中编写代码，实现题目描述的要求。

```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline

"""
请在此处实现代码
"""

x = np.arange(0,10,1)
y1 = x
y2 = x**2
y3 = 0.01*x-0.01

plt.figure(figsize=(10,10))

# 第1个子图
plt.subplot(3, 1, 1)
plt.title("y=x")
plt.plot(x, y1)

# 第2个子图
plt.subplot(3, 1, 2)
plt.title("y=x**2")
plt.plot(x, y2)

# 第3个子图
plt.subplot(3, 1, 3)
plt.title("y=0.01*x-0.01")
plt.plot(x, y3)

plt.show()
```
