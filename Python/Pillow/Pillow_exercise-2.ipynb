### **题目描述**
### 给定一张图片，将其逆时针旋转45度，90度，要求图片内容完整。
提示：rotate()

### **功能实现**

```python
from PIL import Image

"""
请在此处实现代码
"""
img = Image.open("./pillow_data/lena.jpeg",'r')
img = img.resize((400,400))
display(img)
print(img.size)
# 效果如下所示：
```

```python
# 逆时针旋转45度
img45 = img.rotate(45, expand=1)
display(img45)
print(img45.size)
```

```python
# 逆时针旋转90度
img90 = img.rotate(90, expand=1)
display(img90)
print(img90.size)
```
