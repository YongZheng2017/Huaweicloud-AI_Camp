### **题目描述**
### 用pillow库对图片生成缩略图，并加入到原图中。
提示：thumbnail()、paste()

### **功能实现**

```python
import os
from PIL import Image

"""
请在此处实现代码
"""
img = Image.open("./pillow_data/lena.jpeg")
img = img.resize((400, 400))

img_thumbnail = img.copy()
img_thumbnail.thumbnail((80, 80))
img.paste(img_thumbnail, (0,0))

# 效果示例如下：
print("原图：", img.size)
print("缩略图：", img_thumbnail.size)
display(img)
```
