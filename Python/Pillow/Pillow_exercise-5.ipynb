### **题目描述**
### 使用pillow 将图片切割成九宫格，成功霸屏朋友圈。
大家应该都看到过九宫格图片，一张完整的大图被分割成九张小图，在朋友圈和微博里一度成为流行。

实现步骤：
1. 读取图片；
2. 将图片填充为正方形；
3. 切割成九宫格；
4. 保存图片。

效果图如下：

![九宫格](https://modelarts-labs-bj4-v2.obs.cn-north-4.myhuaweicloud.com/course/hwc_edu/python_module_framework/datasets/pillow_data/pic.jpg)

### **功能实现**

```python
import sys
from PIL import Image

# 读取图片
img = Image.open("./pillow_data/lena.jpeg")

# 将图片填充为正方形
img = img.resize((500,500))

# 切割为九宫格
width, height = img.size
item_width = int(width / 3)
box_list = []
for i in range(0, 3):
    for j in range(0, 3):
        box = (j * item_width, i * item_width, (j + 1) * item_width, (i + 1) * item_width)
        box_list.append(img.crop(box))
        
# 保存图片
index = 1
for img in box_list:
        img.save('./pillow_data/' +str(index) + '.png', 'PNG')
        index += 1
        display(img)
```
