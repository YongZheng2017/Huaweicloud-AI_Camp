用 Pillow 为一张图片创建水印,并保存。

如：创建“HELLO WORLD”水印，效果如下图所示：

### **代码实现**

#### 2. 功能实现
请在下面 cell 中编写代码，实现题目描述的要求。

```python
from PIL import Image, ImageDraw, ImageFont

# 读取图片
lena = Image.open("pillow_data/lena.jpeg")

# 使用ImageDraw模块创建水印
draw = ImageDraw.Draw(lena)

# 使用ImageFont创建字体大小
font = ImageFont.truetype(font= "pillow_data/ARIALN.TTF", size= 20)

# 显示
draw.text((20,20),'HELLO WORLD',font=font)
display(lena)

# 保存
lena.save('lena_with_watermark.jpg')
```
