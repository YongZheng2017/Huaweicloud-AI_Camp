### 1、线性回归第三方库实现

```python
# 导入sklearn下的LinearRegression 方法
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
help(LinearRegression())
```

```python
model1 = LinearRegression(fit_intercept=True)  # 请在此处修改 fit_intercept 的参数值
model2 = LinearRegression(fit_intercept=False)

# 构造用于训练的数据集
x_train = np.array([[2, 4], [5, 8], [5, 9], [7, 10], [9, 12]])
y_train = np.array([20, 50, 30, 70, 60])

# 训练模型并输出模型系数和训练结果
model1.fit(x_train, y_train)
model2.fit(x_train, y_train)

# fit(x,y,sample_weight=None)x:训练集 y:目标值 sample_weight:每个样本的个数
# coef_ 系数w,intercept_截距
print(model1.coef_)  # 输出系数w
print(model1.intercept_)  # 输出截距b
print(model1.score(x_train, y_train))  # 输出模型的评估分数R2

print(model2.coef_)  # 输出系数w
print(model2.intercept_)  # 输出截距b
print(model2.score(x_train, y_train))  # 输出模型的评估分数R2
```
